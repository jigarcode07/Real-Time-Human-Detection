```python
import cv2
import numpy as np
from ultralytics import YOLO

# --- SETTINGS ---
input_video =  r"C:\Users\HP\Downloads\13761998_2560_1440_30fps.mp4"  # Replace with your video path
output_video = "outtputt_fixed.mp4"

# --- INITIALIZE ---
cap = cv2.VideoCapture(input_video)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) or 30  # Default to 30 if unavailable

# VideoWriter with reliable settings
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Use 'mp4v' for .mp4 files
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

# Initialize YOLO (optional - remove if not needed)
model = YOLO("yolov8n.pt")

# --- PROCESS VIDEO ---
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # === OPTIONAL: Add your processing here ===
    # Example: YOLO detection (remove if not needed)
    results = model(frame)[0]
    for box in results.boxes:
        if int(box.cls[0]) == 0:  # Only people
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
    # === END PROCESSING ===
    
    # Write frame to output
    out.write(frame)

# --- CLEAN UP ---
cap.release()
out.release()
cv2.destroyAllWindows()
print(f"Success! Video saved to: {output_video}")
```
